```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\nemit\OneDrive\Documents\Python Scripts\pandas\Customer Call List.xlsx")
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df=df.drop(columns='Not_Useful_Column')
```

```python
df
```

```python
df
```

```python
#df["Last_Name"]=df["Last_Name"].str.lstrip('...')
d#f["Last_Name"]=df["Last_Name"].str.lstrip('/')
#df["Last_Name"]=df["Last_Name"].str.rstrip('_')
```

```python
df["Last_Name"]=df["Last_Name"].str.strip('123._/')
```

```python
df
```

```python
df["Phone_Number"].str.strip('-/|')
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
df["Phone_Number"]
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:str(x))
```

```python
df["Phone_Number"]
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:x[0:3]+'-'+x[3:6]+'-'+x[6:10])
```

```python
df["Phone_Number"]
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
```

```python
df[['Street_Address','State','Zipcode']]=df['Address'].str.split(',',n=2,expand=True)
df
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
```

```python
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
```

```python
df
```

```python
#df.replace('N/a','')
#df.replace('NaN','')

df=df.fillna('')
```

```python
df=df.replace('N/a','')
```

```python
df
```

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x,inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x,'Phone_Number']=='':
        df.drop(x,inplace=True)
df
```

```python
df.reset_index(drop=True)
```

```python
df.to_excel(r"C:\Users\nemit\OneDrive\Documents\Python Scripts\pandas\cleaned data.xlsx")
```
